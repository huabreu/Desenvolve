## Identificando Outliers e Removendo-os

```python
dados.boxplot(["Valor"], by = ["Tipo"])
```

```python
grupo_tipo = dados.groupby("Tipo")["Valor"]
```

```python
type(grupo_tipo)
```

```python
grupo_tipo.groups
```

```python
Q1 =  grupo_tipo.quantile(.25)
Q3 =  grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
Q1
```

```python
IIQ
```

```python
limite_superior["Casa"]
```

```python
for tipo in grupo_tipo.groups.keys():
    print(tipo)
```

```python
new_dados = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
    eh_tipo = dados["Tipo"] == tipo
    eh_dentro_limite = (dados["Valor"] >= limite_inferior[tipo]) & (dados["Valor"] <= limite_superior[tipo])
    selecao = eh_tipo & eh_dentro_limite
    dados_selecao = dados[selecao]
    new_dados = pd.concat([new_dados, dados_selecao])
```

```python
new_dados
```

```python
new_dados.boxplot(["Valor"], by = ["Tipo"])
```

```python
new_dados.to_csv("dados/aluguel_residencial_sem_outliers", sep=";", index = False)
```
